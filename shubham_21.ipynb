```python
import pandas as pd

#  Load data
df = pd.read_csv('/content/marketing_campaign.csv', sep='\t')

#  Check column names
print(df.columns)
# Extract columns from the datasets

df_subset = df[['ID', 'Year_Birth', 'Income', 'MntWines', 'MntFruits']]

# Create Age Column if Year_Birth is present
df_subset['Age'] = 2023 - df_subset['Year_Birth']

#  Create new features
df_subset['Total_Spending'] = df_subset['MntWines'] + df_subset['MntFruits']
df_subset['Income_per_Age'] = df_subset['Income'] / df_subset['Age']

#  Display result
print(df_subset.head())
```
